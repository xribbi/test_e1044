```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.datasets import load_iris
from tensorflow.keras.utils import to_categorical
```

```python
iris = load_iris( as_frame=True) # iris data 품질이 좋음 --> 상관관계 높음
iris['frame'] 
```

```python
print( iris['DESCR'] )
```

```python
arr = np.array( iris['frame'].target) 
to_categorical( arr )
```

15분 퀴즈
1. 딥러닝으로 학습
2. 4.9,3.0,1.4, 0.2 에 종분류를 예측하시요

```python
df = iris['frame']

x_data = df.iloc[:,:-1].values
y_data = to_categorical (df.iloc[:,[-1]].values)

```

```python
dense1 = Dense( input_dim =  4, units=100,  activation='relu' )
dense2 = Dense( input_dim =100, units=  3,  activation='softmax' ) 
model = Sequential( [dense1,dense2])
model.compile( loss='categorical_crossentropy', optimizer=Adam(0.1), metrics=['acc'] )
```

```python
h = model.fit( x_data, y_data, epochs=1000)
plt.plot( h.history['loss'])
plt.show()
w,b = dense.get_weights()
print( w )
print( b )

#
# sepal length (cm)	sepal width (cm)	petal length (cm)	petal width (cm) --> target (종분류)
#              4.9,             3.0,                 1.4,               0.2
#
x_quest = np.array(
              [4.9,             3.0,                 1.4,               0.2])

model.predict( x_quest).argmax( axis=1)
```
