```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
- tensorflow 함수를 그래프모드로 변환
- 속도향상
```

```python
@tf.function
def mse( w ):
    hx = w*x # [-1,-1,-1]*[1,2,3] =[-2,-3,-4]
    #c = tf.reduce_mean( (hx-y)**2)
    c = tf.reduce_sum( (hx-y)**2)/len(x) #[-2,-3,-4]-[1,2,3] =[-3,-5,-7]
    return c
```

```python
@tf.function
def gradient( w ):
    # hx = w*x 
    c = tf.reduce_sum( (w*x -y)*2*x )/len(x) 
    return c
```

```python
w = tf.Variable( 10, dtype=tf.float32  ) 
for epoch in range(200):
    c = mse( w )
    g = gradient( w)
    print( 'w=', w.numpy().item(), 
          'mse=',c.numpy().item(), 'gradient=', g.numpy().item() )
    #w = w-0.1*g
    w.assign_sub( 0.1*g )

print('최종=w', w.numpy().item() )
```
