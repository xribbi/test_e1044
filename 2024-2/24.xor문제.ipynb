```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
```

```python
# 공부한시간, 출석일수
x_data = [[0,0],[0,1],[1,0],[1,1]]
y_data = [[0],[1],[1],[0]]
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
initX = tf.initializers.GlorotUniform()

w1 = tf.Variable(  initX(shape=[2,64]) )
b1 = tf.Variable(  initX(shape=[64]) )
w2 = tf.Variable(  initX(shape=[64,32]) )
b2 = tf.Variable(  initX(shape=[32]) )
w3 = tf.Variable(  initX(shape=[32,1]) )
b3 = tf.Variable(  initX(shape=[1]) )


# w1 = tf.Variable(  initX(shape=[2,100]) )
# b1 = tf.Variable(  initX(shape=[100]) )
# w2 = tf.Variable(  initX(shape=[100,1]) )
# b2 = tf.Variable(  initX(shape=[1]) )

# w1 = tf.Variable(  initX(shape=[2,2]) )
# b1 = tf.Variable(  initX(shape=[2]) )
# w2 = tf.Variable(  initX(shape=[2,1]) )
# b2 = tf.Variable(  initX(shape=[1]) )
```

```python
def compute_loss():
    # z1 = tf.matmul(x,w1) + b1
    # hx1 = tf.nn.sigmoid( z1 )
    # z = tf.matmul(hx1,w2 ) + b2
    # hx = tf.nn.sigmoid( z )
    z1 = tf.matmul(x,w1) + b1
    hx1 = tf.nn.sigmoid( z1 )
    z2 = tf.matmul(hx1,w2 ) + b2    
    hx2 = tf.nn.sigmoid( z2 )
    z = tf.matmul(hx2,w3) + b3
    
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits( labels=y, logits=z)
    cost = tf.reduce_mean( cost_i)
    return cost
```

```python
def hxfn( xd ):
    xd = tf.constant( xd, tf.float32)
    # z1 = tf.matmul(xd,w1) + b1
    # hx1 = tf.nn.sigmoid( z1 )
    # z = tf.matmul(hx1,w2 ) + b2
    # hx = tf.nn.sigmoid( z )
    z1 = tf.matmul(xd,w1) + b1
    hx1 = tf.nn.sigmoid( z1 )
    z2 = tf.matmul(hx1,w2 ) + b2    
    hx2 = tf.nn.sigmoid( z2 )
    z = tf.matmul(hx2,w3) + b3    
    hx =tf.nn.sigmoid( z)
    return (hx.numpy() > 0.5) +0
```

```python
optimizer = Adam( learning_rate=0.01)
for i in range(200):
    with tf.GradientTape() as tape:
        cost = compute_loss() 
    gradients = tape.gradient( cost, [w1,w2,w3,b1,b2,b3] )
    optimizer.apply_gradients(  zip( gradients, [w1,w2,w3,b1,b2,b3] ) )
    
    pred = hxfn( x_data)
    acc = accuracy_score(y_data, pred)
    print(i, 'cost:', cost.numpy(),'정확도', acc )
```
