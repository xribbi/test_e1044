```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
df = pd.read_csv('data/pima-indians-diabetes.data.csv')
df
```

```python
x_data = df.iloc[:, :-1].values
y_data = df.iloc[:, [-1] ].values
```

```python
x_data.shape
```

```python
x_train, x_test, y_train, y_test = train_test_split( x_data, y_data, 
                                            test_size=0.3, stratify=y_data)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
dense = Dense( units=1, input_dim=8, activation='sigmoid')
model = Sequential( [dense])
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_train, y_train, epochs=500)
```

### train 예측값

```python
predT = (model.predict( x_train) > 0.5) +0
predT
```

```python
accuracy_score( y_train, predT )
```

### test 예측값

```python
predTest = (model.predict( x_test) > 0.5) +0
predTest
```

```python
accuracy_score( y_test, predTest )
```
