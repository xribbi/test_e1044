```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
def mse( x,y,w):
    c = 0
    for i in range( len(x)):
        hx = w*x[i]
        c = c+ (hx-y[i])**2
    return c/len(x)
```

```python
def gradient( x,y,w):
    c = 0
    for i in range( len(x)):
        hx = w*x[i]
        c = c+ (hx-y[i])*x[i]
    return c/len(x)
```

```python
x_data = [1,2,3]
y_data = [1,2,3]
```

```python
w = 10
for epoch in range(200):
    c = mse( x_data, y_data, w )
    g = gradient( x_data, y_data, w)
    print( 'w=', w, 'mse=',c, 'gradient=', g )
    w = w-0.1*g

print('최종=w', w )
```
