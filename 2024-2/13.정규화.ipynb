```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import MinMaxScaler, StandardScaler
```

```python
data = [[828, 920, 1234567, 1020, 1111],
            [824, 910, 2345612, 1090, 1234],
            [880, 900, 3456123, 1010, 1000],
            [870, 990, 2312123, 1001, 1122],
            [860, 980, 3223123, 1008, 1133],
            [850, 970, 2432123, 1100, 1221]]
data = np.float32(data)
```

```python
df = pd.DataFrame( data )
df
```

```python
x_data = df.iloc[:, :-1].values
y_data = df.iloc[:, [-1]].values
```

```python
# 3.14e+02
3.14e-2
```

```python
scaleF = MinMaxScaler()
x_dataN = scaleF.fit_transform( x_data )
```

```python
scaleL = MinMaxScaler()
y_dataN = scaleL.fit_transform( y_data )
```

```python
dense = Dense( units=1, input_dim=4)# w:3x1, b:1
model = Sequential( [dense] )
model.compile( loss='mse', optimizer= Adam(0.1) )
h = model.fit( x_dataN, y_dataN, epochs= 500)
```

```python
#828, 920, 1234567, 1020 예측
xd =scaleF.transform( [[828, 920, 1234567, 1020]] )
xd
```

```python
pred = model.predict( xd )
pred
```

```python
scaleL.inverse_transform( pred )
```
