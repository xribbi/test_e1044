```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
# 공부한시간, 출석일수
x_data =np.array( [[1,3],[2,2],[3,1],[4,6],[5,5],[6,2],[1,1],[2,1],
          [2,1],[4,5],[5,6],[6,7]] )
y_data =np.array( [[0],[0],[0],[1],[1],[1],[0],[0],[0],[1],[1],[1]] )
```

```python
x_train, x_val, y_train, y_val = train_test_split( x_data, y_data, 
                                    test_size=0.2, stratify=y_data)
```

```python
x = tf.constant( x_train, tf.float32)
y = tf.constant( y_train, tf.float32)
# xavier 초기화
initX = tf.initializers.GlorotUniform()
w = tf.Variable(  initX(shape=[2,1]) )
b = tf.Variable(  initX(shape=[1]) )
```

```python
def compute_loss():
    z = tf.matmul(x,w) + b
    #hx = tf.nn.sigmoid( z ) #0~1사이 확률값
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits( labels=y, logits=z)
    cost = tf.reduce_mean( cost_i)
    return cost
```

```python
def hxfn( xd ):
    xd = tf.constant( xd, tf.float32)
    z = tf.matmul( xd,w ) +b
    hx = tf.nn.sigmoid(z)
    return (hx.numpy() > 0.5) +0
```

```python
x_train.shape
```

```python
optimizer = Adam( learning_rate=0.01)
for i in range(500):
    with tf.GradientTape() as tape:
        cost = compute_loss() 
    gradients = tape.gradient( cost, [w,b] )
    optimizer.apply_gradients(  zip( gradients, [w,b] ) )
    
    pred = hxfn( x_train)
    acc = accuracy_score(y_train, pred)
    predV = hxfn( x_val)
    accV = accuracy_score(y_val, predV)
    print(i, 'cost:', cost.numpy(),
           '정확도', acc, 'valid acc', accV )
```
