```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0]] )
```

```python
dense = Dense( units=1, input_dim=2, activation='sigmoid')
model = Sequential( [dense])
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=500)
```

```python
dense1 = Dense( units=100, input_dim=2, activation='sigmoid')# w:2x100, b:100
dense2 = Dense( units=1, input_dim=100, activation='sigmoid')# w:100x1 ,b:1
model = Sequential( [dense1,dense2])
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=500)
```
