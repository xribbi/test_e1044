```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential, Model
from tensorflow.keras.layers import Dense, Input
```

```python
x_data = np.array( [[0,0],[0,1],[1,0],[1,1]] )
y_data = np.array( [[0],[1],[1],[0]] )
```

### 1.케라스모델

```python
dense1 = Dense( units=64, input_dim=2, activation='relu')
dense2 = Dense( units=1, input_dim=64, activation='sigmoid')
model = Sequential( [dense1, dense2])
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=500)
```

### 2.케라스 모델

```python
model = Sequential()
model.add( Dense( units=64, input_dim=2, activation='relu') )
model.add( Dense( units=1, input_dim=64, activation='sigmoid') )
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=300)
```

### 3.케라스모델

```python
input = Input( shape=(2,))
x = Dense( units=64, activation='relu')(input)
x = Dense( units=32, activation='relu')(x)
pred = Dense( units=1, activation='sigmoid')(x)

model = Model( inputs=input, outputs=pred)
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=300)
```

### 4.케라스모델

```python
class XorModel( Model):
    def __init__(self):
        super(XorModel, self).__init__()
        self.dense1 =Dense(units=64, activation='relu')
        self.dense2 =Dense(units=32, activation='relu')
        self.dense3 =Dense(units=1, activation='sigmoid')
    def call(self, xd):
        x = self.dense1( xd )
        x = self.dense2( x )
        hx = self.dense3( x )
        return hx
```

```python
model = XorModel()
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=300)
```
