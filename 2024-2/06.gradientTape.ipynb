```python
x = tf.constant([1,2,3,4,5], tf.float32)
y = tf.constant([3,5,7,9,11], tf.float32)
w = tf.Variable( 10, dtype=tf.float32)
b = tf.Variable( 10, dtype=tf.float32)
```

```python
@tf.function
def compute_loss():
    hx = w*x + b
    c = tf.reduce_mean( (hx-y)**2)
    return c 
```

```python
optimizer = SGD( learning_rate=0.01)
for i in range(2000):
    with tf.GradientTape() as tape:
        cost = compute_loss() #y = tf.reduce_mean( (w*x+b-y)**2)
    gradients = tape.gradient( cost, [w,b] )
    optimizer.apply_gradients(  zip( gradients, [w,b] ) )
    # 
    #w.assign_sub( 0.01*w)
    #b.assign_sub( 0.01*b)
    print(i, 'cost:', cost.numpy() )
```
