```python
import tensorflow as tf
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = tf.constant([1,2,3,4,5], tf.float32)
y = tf.constant([3,5,7,9,11], tf.float32)
```

```python
w = tf.Variable( 10, dtype=tf.float32)
b = tf.Variable( 10, dtype=tf.float32)
epochs = 2000
learning_rate = 0.01
n = len(x )

for i in range( epochs):
    hx = w*x + b
    cost = tf.reduce_sum( (hx-y)**2)/n
    gradientW = tf.reduce_sum( (w*x+b-y)*2*x)/n
    gradientB = tf.reduce_sum( (w*x+b-y)*2)/n
    print('cost', cost.numpy().item(),'w=',w.numpy().item(),
          'b=',b.numpy().item(),
          'gw=',gradientW.numpy().item(),
           'gb=',gradientB.numpy().item() )
    w.assign_sub( learning_rate*gradientW)
    b.assign_sub( learning_rate*gradientB)
print('w', w.numpy().item(), 'b', b.numpy().item() )
```

```python
x1 = tf.constant([1,2,3,4,5], tf.float32)
x2 = tf.constant([1,2,3,4,5], tf.float32)
y = tf.constant([3,5,7,9,11], tf.float32)

w1 = tf.Variable( 10, dtype=tf.float32)
w2 = tf.Variable( 10, dtype=tf.float32)
b = tf.Variable( 10, dtype=tf.float32)
epochs = 2000
learning_rate = 0.01
n = len(x )

for i in range( epochs):
    hx = w1*x1 + w2*x2 + b
    cost = tf.reduce_sum( (hx-y)**2)/n
    gradientW1 = tf.reduce_sum( (w1*x1 + w2*x2 + b-y)*2*x1)/n
    gradientW2 = tf.reduce_sum( (w1*x1 + w2*x2 + b-y)*2*x2)/n
    gradientB = tf.reduce_sum( (w1*x1 + w2*x2 + b-y)*2)/n
    print('cost', cost.numpy().item(),'w=',w.numpy().item(),
          'b=',b.numpy().item(),
          'gw=',gradientW.numpy().item(),
           'gb=',gradientB.numpy().item() )
    w1.assign_sub( learning_rate*gradientW1)
    w2.assign_sub( learning_rate*gradientW2)
    b.assign_sub( learning_rate*gradientB)
print('w', w.numpy().item(), 'b', b.numpy().item() )
```
