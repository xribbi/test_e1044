```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
x_data = np.array( [[1,3],[2,2],[3,1],[4,6],[5,5],[6,2]] )
y_data = np.array( [[0],[0],[0],[1],[1],[1]] )
```

```python
# z = tf.matmul( x, w ) + b
# hx = sigmoid( z )
```

```python
# dense = Dense( units=1, input_dim=2, activation='sigmoid')# w:2x1, b=1
# dense( tf.constant( x_data, tf.float32) )
```

```python
dense = Dense( units=1, input_dim=2, activation='sigmoid')
model = Sequential( [dense])
model.compile( loss='binary_crossentropy', optimizer=Adam(0.1),
              metrics=['acc'] )
h = model.fit( x_data, y_data, epochs=500)
```

```python
# dense( tf.constant( x_data, tf.float32) ).numpy() >0.5
dense( tf.constant( x_data, tf.float32) )
```

```python
pred = ( model.predict( x_data) > 0.5) + 0
pred
```

```python
(pred == y_data).mean()
```

```python
from sklearn.metrics import accuracy_score, f1_score, confusion_matrix
```

```python
accuracy_score( y_data, pred)
```

```python
f1_score( y_data, pred )
```

```python
confusion_matrix( y_data, pred)
```
