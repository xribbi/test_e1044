```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
```

```python
x1_data = [1,0,3,4,5] # 공부한 시간
x2_data = [0,1,2,3,4] # 출석일수
y_data = [1,2,3,4,5] # 점수
```

```python
x1 = tf.constant(x1_data, tf.float32)
x2 = tf.constant(x2_data, tf.float32)
y = tf.constant(y_data, tf.float32)
w1 = tf.Variable( tf.random.uniform([1]), dtype=tf.float32)
w2 = tf.Variable( tf.random.uniform([1]), dtype=tf.float32)
b = tf.Variable( tf.random.uniform([1]), dtype=tf.float32)
```

```python
@tf.function
def compute_loss():
    hx = w1*x1+ w2*x2 + b
    c = tf.reduce_mean( (hx-y)**2)
    return c 
```

```python
optimizer = Adam( learning_rate=0.01)
for i in range(2000):
    with tf.GradientTape() as tape:
        cost = compute_loss() #y = tf.reduce_mean( (w*x+b-y)**2)
    gradients = tape.gradient( cost, [w1,w2,b] )
    optimizer.apply_gradients(  zip( gradients, [w1,w2,b] ) )
    print(i, 'cost:', cost.numpy() )
```

```python
def hxfn( xd1, xd2):
   hx = w1*xd1+ w2*xd2 + b
   return hx.numpy()
```
