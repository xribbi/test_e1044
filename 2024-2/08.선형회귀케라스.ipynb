```python
import tensorflow as tf
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from tensorflow.keras.optimizers import Adam
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

- 텐서:로우레벨, 응용 
- 케라스:하이레벨

```python
df = pd.read_csv('data/cars.csv', index_col='Unnamed: 0')
df
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
#1.w와b: w 1, b 1
#2.hx = w*x + b
dense = Dense( units=1, input_dim=1 )
model = Sequential( [dense] )
model.compile( loss='mse', optimizer=Adam(0.1) )
model.fit( x_data, y_data, epochs=500)
```

```python
w, b = dense.get_weights()
print(w )
print( b )
```

```python
model.predict( np.array( [10] ) ) # 1.텐서변화 2. dense( x ) 3. numpy()
```
