```python
df = pd.read_csv('data/titanic.csv')
df['ngen']=df['Sex'].apply(lambda v: 0 if v=='male' else 1)
df['Age'].isna().sum()
```

```python
df['Age'].fillna(df['Age'].mean(), inplace=True)
df
```

```python
x_data = df[['Fare', 'Pclass', 'ngen', 'Age']]
y_data = df['Survived']

print(x_data.shape, y_data.shape)
```

```python
model = Sequential()
model.add(Dense(units=32, input_dim=4, activation='relu'))
model.add(Dense(units=16, input_dim=32, activation='relu'))
model.add(Dense(units=1, input_dim=16, activation='sigmoid'))
model.compile(loss='binary_crossentropy', optimizer=Adam(0.01), metrics=['acc'])
model.fit(x_data, y_data, epochs=300)
```

```python
# 'Fare'=30, 'Pclass'=1, 'ngen'=1, 'Age'=30
[30, 1, 1, 30]

model.predict([[30, 1, 1, 30]])
```
