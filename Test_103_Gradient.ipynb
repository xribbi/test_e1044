```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
## 일반적으로 Mean Square Error 
## 특별하게는 Mean Absolute Error
## MSE, MAE 모두 Cost 로 정의 가능

x_data = [1.0,2,3]
y_data = [1.0,2,3]
```

```python
plt.scatter(x_data,y_data);plt.xlim(0,5);plt.ylim(0,5)
```

```python
def mse(x,y,w, b=0.0):
    sum = 0
    for i in range(len(x)):
        hx = w*x[i]+b
        sum = sum + (hx-y[i])**2
    return sum/len(x)

def mse_grad(x,y,w, b=0.0):
    sum = 0
    for i in range(len(x)):
        hx = w*x[i]+b
        sum = sum + (hx-y[i])*x[i]
    return sum/len(x)
```

```python
w  = 10.0 ; # (random) initial w
lr =  0.1 ; # learning rate
for epoch in range(200):
    c = mse(x_data, y_data, w)
    g = mse_grad(x_data, y_data, w)
    print("w =%8.4f, mse = %8.4f, mse_grad= %8.4f"% (w,c,g))
    w = w - lr * g

print("w (final) =",w)

```

```python
print(mse(x_data,y_data,-1))
print(mse(x_data,y_data, 0))
print(mse(x_data,y_data, 1))
print(mse(x_data,y_data, 2))
print(mse(x_data,y_data, 3))
```

```python
for w in np.linspace(-3,5,50):
    cost=mse(x_data,y_data,w)
    print("%8.4f,%8.4f"%(w,cost))
    plt.plot(w,cost,'ro')
plt.xlabel('w')
plt.ylabel('mse')
```
