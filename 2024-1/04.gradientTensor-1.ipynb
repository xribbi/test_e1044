```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1,2,3] 
y_data = [1,2,3]
x = tf.constant( x_data, dtype=tf.float32)
y = tf.constant( y_data, dtype=tf.float32)
```

```python
def mse( x,y, w):
    hx = w*x # [-1,-1,-1] *[1,2,3] = [-1,-2,-3]
    # hx-y : [-1,-2,-3]- [1,2,3 ]  = [-2,-4,-6] = [4,16,36]
    c = tf.reduce_sum( (hx-y)**2 )/len(x)
    return c.numpy()
```

```python
def gradient( x,y, w):
    # hx = w*x
    c = tf.reduce_sum( (w*x-y)*x*2 )/len(x)
    return c.numpy()
```

```python
print( mse( x, y, -1))
print( mse( x, y, 0))
print( mse( x, y, 1))
print( mse( x, y, 2))
```

```python
w = 10
for epoch in range( 200):
    c = mse( x, y, w )
    g = gradient( x, y, w )
    print( 'w=',w, 'gradient=',g, 'mse=', c )
    w = w-0.1*g
print( '최종w:', w )
```
