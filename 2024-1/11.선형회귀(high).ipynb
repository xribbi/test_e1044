```python
import tensorflow as tf
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('data/cars.csv' ,index_col='Unnamed: 0')
df
```

```python
x_data = df['speed'].values
y_data = df['dist'].values
```

```python
x_data.reshape( -1, 1)
```

```python
x_data.shape
```

```python
dense = Dense( units=1, input_dim=1 )
dense( x_data.reshape( -1, 1) )
```

```python
w, b = dense.get_weights()
print( w, b )
```

```python
dense = Dense( units=1, input_dim=1 )
model = Sequential( [dense] )
model.compile( loss='mse', optimizer=Adam( 0.1 ))
h = model.fit( x_data, y_data, epochs= 500 )
```

```python
w, b = dense.get_weights()
print( w, b )
```

```python
# dense( x_data.reshape( -1, 1) )
pred = model.predict( x_data )
```

```python
plt.scatter( x_data, y_data )
plt.plot( x_data, pred, 'r--' )
plt.show()
```

```python
plt.plot( h.history['loss'] )
plt.show()
```
