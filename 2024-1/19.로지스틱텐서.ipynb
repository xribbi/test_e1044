```python
import tensorflow as tf
from tensorflow.keras.optimizers import Adam
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x_data = [[1,3],[2,2],[3,1],[4,6],[5,5],[6,4]] #공부한시간 ,출석일수
y_data = [[0],[0],[0],[1],[1],[1]]
```

```python
x = tf.constant( x_data, tf.float32)
y = tf.constant( y_data, tf.float32)
```

```python
initX = tf.initializers.GlorotUniform() # xavier 알고리즘에 의한 최적 초기화(랜덤)
w = tf.Variable( initX( shape=[2,1] ) )
b = tf.Variable( initX( shape=[1] ) )
```

```python
def hxfn( xdata ):
    xd = tf.constant( xdata, tf.float32)
    z = tf.matmul( xd,w) + b
    hx = tf.nn.sigmoid( z )
    return (hx.numpy() > 0.5) +0
    # return z.numpy(), hx.numpy()
```

```python
def accFn():
    p = hxfn( x_data )
    print('p=', p)
    print( '------------------')
    return (y_data==p).mean()
```

```python
def compute_loss():
    z = tf.matmul( x, w) + b
    # hx = tf.nn.sigmoid( z )
    cost_i = tf.nn.sigmoid_cross_entropy_with_logits( labels=y, logits=z )
    c  = tf.reduce_mean( cost_i )
    return c
```

```python
optimizer = Adam( 0.1 )
for i in range( 1000):
    optimizer.minimize( compute_loss, var_list=[w,b])
    c = compute_loss().numpy()
    print( i, 'cost=',c ,'정확도:', accFn() )
```
