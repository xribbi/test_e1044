```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x_data = [1,2,3] 
y_data = [1,2,3]
```

```python
def mse( x,y,w):
    c = 0
    for i in range( len(x)):
        hx = w*x[i]
        c = c+(hx-y[i])**2
    return c/len(x)
```

```python
def gradient( x,y,w):
    c = 0
    for i in range( len(x)):
        hx = w*x[i]
        c = c+(hx-y[i])*x[i]
    return c/len(x)
```

```python
w = 10
hist =[]
for epoch in range(200):
    c = mse( x_data, y_data, w )
    g = gradient( x_data, y_data, w )
    hist.append( c )
    print( 'w=',w, 'gradient=',g, 'mse=', c )
    w = w-0.1*g
print( w )
```

```python
plt.plot( hist )
plt.show()
```
