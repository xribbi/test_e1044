```python
import tensorflow as tf
from tensorflow.keras.optimizers import SGD
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = tf.constant( [1,2,3,4,5], tf.float32)
y = tf.constant( [3,5,7,9,11], tf.float32)
```

```python
w = tf.Variable( 10.0, tf.float32)
b = tf.Variable( 10.0, tf.float32)
```

```python
def compute_loss():
    hx = w*x + b
    c = tf.reduce_mean( (hx-y)**2)
    return c
```

```python
optimizer = SGD( learning_rate=0.01 )
for i in range(2000):
    optimizer.minimize( compute_loss, var_list=[w,b] )
    # w = w- 0.01*w의기울기  #w.assign_sub( 0.01*w의기울기 )
    # b = b- 0.01*b의기울기 ##b.assign_sub( 0.01*b의기울기 )
    c = compute_loss().numpy()
    print( i, 'w=',w.numpy(), 'cost=', c )
```
