```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = tf.constant( [1,2,3,4,5], tf.float32)
y = tf.constant( [3,5,7,9,11], tf.float32)
```

```python
plt.ylim( 0,12 )
plt.xlim( 0,7 )
plt.scatter( x, y)
plt.show()
```

```python
w = 10
b = 10
epochs = 2000
learning_rate = 0.01
n = len(x)
```

```python
for i in range( epochs ):
    hx = w*x + b
    c = tf.reduce_sum( (hx-y)**2)/n
    gradientW = tf.reduce_sum( (w*x + b-y)*2*x)/n
    gradientB = tf.reduce_sum( (w*x + b-y)*2 )/n
    w = w-learning_rate*gradientW
    b = b-learning_rate*gradientB
    print( 'w=',w.numpy(),'b=',b.numpy(), 'gradientW=',gradientW.numpy(),
          'gradientB=',gradientB.numpy(), 'mse=', c.numpy() )    
print( ':w',w.numpy(), ':b', b.numpy() )
```
