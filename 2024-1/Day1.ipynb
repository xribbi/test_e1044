```python
1. y = (x+3)^3의 미분과정

t=(x+3) 이면 y=t^3
dy/dx = dy/dt*dt/dx = (3t)^2*(1) = 3(x+3)^2

2.
x = [1,2,3,4,5,6,7]
y = [5,8,11,14,17,20,23]
인경우 w 와 b를 구하시요
직접구현과, 텐서라이브러리 사용버전
x= 10 인경우 예측값을 구하시요
```

```python
# 직접 구현
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt

x = tf.constant([1,2,3,4,5,6,7], tf.float32)
y = tf.constant([5,8,11,14,17,20,23], tf.float32)

w = 10
b = 10
epoch = 2000
learning_rate = 0.01
n = len(x)

for i in range(epoch):
  hx = w * x + b
  c = tf.reduce_sum((hx-y)**2)/n
  gradientW = tf.reduce_sum((w * x + b -y)*2*x)/n
  gradientB = tf.reduce_sum((w * x + b -y)*2)/n

  w = w - learning_rate * gradientW
  b = b - learning_rate * gradientB

  print('w=', w.numpy(), 'b=', b.numpy(), 'gradientW=',gradientW.numpy(), 'gradientB=', gradientB.numpy(), 'mse=', c.numpy())
print('최종 W =', w.numpy(), '최종 B =', b.numpy())
```

```python
# 텐서라이브러리 사용버전

import tensorflow as tf
from tensorflow.keras.optimizers import SGD
import numpy as np
import matplotlib.pyplot as plt

x = tf.constant([1,2,3,4,5,6,7], tf.float32)
y = tf.constant([5,8,11,14,17,20,23], tf.float32)

# x = tf.constant([1,2,3,4,5], tf.float32)
# y = tf.constant([3,5,7,9,11], tf.float32)

w = tf.Variable(10.0, tf.float32)
b = tf.Variable(10.0, tf.float32)

def compute_loss():
  hx = w*x + b
  c = tf.reduce_mean((hx-y)**2)
  return c

optimizer = SGD(learning_rate=0.01)
for i in range(2000):
  optimizer.minimize(compute_loss, var_list=[w,b])
  c = compute_loss().numpy()
  print(i, 'w=', w.numpy(), 'cost=', c)
```
